```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url_list = []
prices_list = []
propTitles = []
propValues = []

for i in range(1,29): #29 sayfa veri alıcak oda yaklaşık 600-700 arası bir veri demek
    url = "https://www.trendyol.com/cep-telefonu?pi=" + str(i)
    r = requests.get(url)
    source = BeautifulSoup(r.content, "lxml") #istek yaptığımız sayfanın içeriğini çekiyoruz
    
    urls = source.find_all("div",attrs={"class":"p-card-chldrn-cntnr"}) #tüm verilerin bulunması
    for url in urls:
        url_phone = "https://www.trendyol.com/"+url.a.get("href")
        url_list.append(url_phone) #buraya kadar verileri ve linklerini bulduk
        print(url_phone)
        
        r_phone = requests.get(url_phone)
        source_phone = BeautifulSoup(r_phone.content,"lxml")
        
        properties = source_phone.find_all("div", attrs={"class":"prop-item"})
        for prop in properties:
            prop_title = prop.find("div",attrs={"class":"item-key"}).text
            prop_value = prop.find("div",attrs={"class":"item-value"}).text
            propTitles.append(prop_title)
            propValues.append(prop_value)
            
    prices = source.find_all("div",attrs={"class":"prc-box-sllng"}) #fiyat özelliklerinini bulunması
    for price in prices:
        prices_list.append(price.text) #fiyatlar üzerinde gezinme işlemi ve listeye yazma
        print(price.text)
        
print(str(len(url_list))+ " Adet link bulundu.")
print(str(len(prices_list))+ " Adet fiyat bulundu.")
print(str(len(propTitles))+ " Adet özellik başlığı bulundu.")
print(str(len(propValues))+ " Adet özellik verisi bulundu.")
```

```python
#excelden önce fiyatları bir data frame yazdırıyoruz
df_urls = pd.DataFrame()
df_urls["urls"] = url_list
df_urls["prices"] = prices_list
```

```python
phones = len(url_list) #bulduğumuz veri sayısı
```

```python
columns = np.array(propTitles)
columns = np.unique(columns) #bulunan özelliklerin unique halini buluyor 
```

```python
df = pd.DataFrame(columns=columns)
df["url"] = url_list
df["price"] = prices_list
```

```python
df.head() #özeliiklere bakıyoruz
```

```python
#dataframe ile bütün verileri tek tek çekip dataframedeki sütünlara yazdırma işlemi
for i in range(0,phones):
    url = df['url'].loc[i]
    r = requests.get(url)
    source = BeautifulSoup(r.content,"lxml")
    
    properties = source.find_all("div",attrs={"class":"prop-item"})
    for prop in properties:
        prop_title = prop.find("div",attrs={"class":"item-key"}).text
        prop_value =prop.find("div",attrs={"class":"item-value"}).text
        print(prop_title+prop_value)
        df[prop_title].loc[i] = prop_value
```

```python
#verileri doldurmadan önce bir excel dosyasına alıyoruz
df.to_excel("./Desktop/verilerin_ilk_hali.xlsx", index=False)
```
